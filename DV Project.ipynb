```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
```

Problem Definition :- We are trying to find which manufracture gets highest rating

```python
cereal.info()
```

```python
cereal.describe()
```

as we can see in describe that 3 columns has negative value so first we are going to change that then we will deal with null values

```python
cereal["carbo"].replace({-1: 15.00}, inplace=True)
cereal["sugars"].replace({-1: 7.00}, inplace=True)
cereal["potass"].replace({-1: 93.00}, inplace=True)
```

```python
cereal.isna().sum()
```

```python
cereal.describe()
```

```python
cereal['mfr'].fillna(method='ffill',inplace=True)#propagate[s] last valid observation forward to next valid

cereal['type'].fillna(method='bfill',inplace=True)#works same as ffill but in backward direction

cereal['calories'].fillna(105.5,inplace=True)

cereal['protein'].fillna(2.5,inplace=True)

cereal['fat'].fillna(1,inplace=True)

cereal['sodium'].fillna(161.00,inplace=True)

cereal['fiber'].fillna(2.0,inplace=True)

cereal['carbo'].fillna(14.0,inplace=True)

cereal['sugars'].fillna(7.00,inplace=True)

cereal['potass'].fillna(93.00,inplace=True)

cereal['vitamins'].fillna(28.00,inplace=True)

cereal['shelf'].fillna(2.0,inplace=True)

cereal['weight'].fillna(1.0,inplace=True)

cereal['cups'].fillna(0.81,inplace=True)

cereal['rating'].fillna(42.87,inplace=True)
```

```python
cereal.describe()
```

# 1)

```python
cereal_mfr=cereal.groupby('mfr').mean().reset_index()

print('      Manufacture V/S Rating')
fig1=sns.barplot(x='mfr',y='rating',data=cereal_mfr,palette='twilight_shifted',ci=None,edgecolor='k')
for index,row in cereal_mfr.iterrows():
    fig1.text(row.name,row.rating,round(row.rating,2),color='k',ha='center',size=17)
plt.show()
```

Thus the Nabisco manufacture has highest rating of 67.97
(avg) while German Mills has lowest rating of 34.49(avg)
